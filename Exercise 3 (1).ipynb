````python
# TODO: ูุชุฏ ุฒุฑ ุฑุง ุจุฑุง ูุดุงู ุฏุงุฏู ูุณุชูฺฏุฑุงู ุชฺฉูู ู ูุชุบุฑ ูุง ุฐฺฉุฑ ุดุฏู ุฑุง ุจุฑุฑุณ ฺฉูุฏ.
def hist_plot(df, numeric_columns,
              bins=30, rot=None,
              figsize=(14, 8)):
    """
    ุงู ุชุงุจุน ฺูุฏู ูุชุบุฑ ุนุฏุฏ ุฑุง ุจู ุตูุฑุช ูุณุชูฺฏุฑุงู ููุงุด ูโุฏูุฏ.
    ูพุงุฑุงูุชุฑูุง:
    df : DataFrame -> ุฏุชุงูุฑู ูุฑูุฏ
    numeric_columns : list -> ูุณุช ูุชุบุฑูุง ุนุฏุฏ
    bins : int -> ุชุนุฏุงุฏ ุณุทูโูุง (bins) ูููุฏุงุฑ
    rot : int -> ููุฏุงุฑ ฺุฑุฎุด ุจุฑฺุณุจโูุง ูุญูุฑ X
    figsize : tuple -> ุงูุฏุงุฒู ูููุฏุงุฑ
    """
    n_cols = 2  # ุชุนุฏุงุฏ ุณุชููโูุง subplot
    n_rows = math.ceil(len(numeric_cols) / n_cols) 
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(figsize[0], figsize[1] * 2))
    axes = axes.flatten()  
    for i,col in enumerate(numeric_cols):
        df[col] = pd.to_numeric(df[col], errors='coerce')
        ax=axes[i]
        sns.histplot(df[col], bins=bins,kde=True,ax=ax, legend=True, color="skyblue"), 
        # sns.histplot(df[col].dropna(), bins=bins,kde=True,ax=ax, legend=True, color="skyblue"), 
        ax.set_title(f'Histogram of {col}\nMean: {df[col].mean():.2f}, Std: {df[col].std():.2f}')
        ax.set_xlabel(col)
        ax.set_ylabel('Frequency')
        ax.tick_params(axis='x', labelrotation=rot)
    for j in range(i+1, len(axes)):
        fig.delaxes(axes[j])
    
    plt.tight_layout()
    plt.show()
````

````python
# TODO: ูุชุฏ ุฒุฑ ุฑุง ุจุฑุง ูุดุงู ุฏุงุฏู KDE Plot ุชฺฉูู ู ูุชุบุฑ ูุง ุฐฺฉุฑ ุดุฏู ุฑุง ุจุฑุฑุณ ฺฉูุฏ.
def plot_numeric_distributions(df, numeric_columns, target_col='Churn Label'):
    """
    ุฑุณู ูููุฏุงุฑูุง KDE ุจุฑุง ูุชุบุฑูุง ุนุฏุฏ ุจุง ุชูฺฉฺฉ ุจุฑ ุงุณุงุณ ูุชุบุฑ ูุฏู.
    ุงู ุชุงุจุน ุชูุฒุน ูุชุบุฑูุง ุนุฏุฏ ุฑุง ุจุฑุง ูุฑ ููุฏุงุฑ ุงุฒ ูุชุบุฑ ูุฏู (ูุงููุฏ 'Churn Label')
    ุฏุฑ ฺฉ ุดุจฺฉู ฒรฒ ููุงุด ูโุฏูุฏ.
    ูพุงุฑุงูุชุฑูุง:
    df (DataFrame): ุฏุชุงูุฑู ุดุงูู ุฏุงุฏูโูุง
    numeric_cols (list): ูุณุช ุณุชููโูุง ุนุฏุฏ ุจุฑุง ุฑุณู ูููุฏุงุฑ
    target_col (str): ูุงู ุณุชูู ูุฏู ุจุฑุง ุฏุณุชูโุจูุฏ (ูพุดโูุฑุถ 'Churn Label')
    """
    n_cols = 2  # ุชุนุฏุงุฏ ุณุชููโูุง subplot
    n_rows = math.ceil(len(numeric_cols) / n_cols) 
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(10,8))
    axes = axes.flatten() 
    palette = {'No': 'blue', 'Yes': 'red'}

    # ุฑุณู ูุฑ ฺฉุฏุงู ุงุฒ ุณุชููโูุง ุนุฏุฏ
    for i, col in enumerate(numeric_cols):
        df[col] = pd.to_numeric(df[col], errors='coerce')
        ax = axes[i]
        sns.kdeplot(data=df, x=df[col], hue=target_col, ax=ax, fill=True, common_norm=False,warn_singular=False, palette=palette)
        ax.set_title(f'Distribution of {col} by {target_col}\nMissing: {df[col].isna().sum()}')
        ax.set_xlabel(col)
        ax.set_ylabel('Density')
        # ุงูุฒูุฏู ูุงูฺฏู ุจุฑุง ูุฑ ฺฏุฑูู
        for hue_val in df[target_col].unique():
            mean_val = df[df[target_col] == hue_val][col].mean()
            ax.axvline(mean_val, linestyle='--', label=f'Mean {hue_val}: {mean_val:.2f}', alpha=0.5)
        ax.legend()


    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()
````

````python
# TODO: ูุชุฏ ุฒุฑ ุฑุง ุชฺฉูู ฺฉูุฏ ุจุฑุง ุจุฑุฑุณ ุงุฑุชุจุงุท ูุชุบุฑูุง ุฏุณุชูโุง ุจุง  ุจุง ูุฏู.
def plot_categorical_subplots(df,
                              categorical_cols,
                              target_col='Churn Label'):
    """
    ุฑุณู ูููุฏุงุฑูุง Countplot ุจุฑุง ูุชุบุฑูุง ุฏุณุชูโุง ููุฑุงู ุจุง ูุชุบุฑ ูุฏู.
    ุงู ุชุงุจุน ูููุฏุงุฑูุง ุดูุงุฑุด ุฑุง ุจุฑุง ูุฑ ูุชุบุฑ ุฏุณุชูโุง ูุดุฎุตโุดุฏู ุฏุฑ ฺฉูุงุฑ ูุชุบุฑ ูุฏู
    (ูุงููุฏ 'Churn Label') ุฏุฑ ฺฉ ุดุจฺฉู ฒรฒ ููุงุด ูโุฏูุฏ.
    ูพุงุฑุงูุชุฑูุง:
    df (DataFrame): ุฏุชุงูุฑู ุดุงูู ุฏุงุฏูโูุง
    categorical_cols (list): ูุณุช ุณุชููโูุง ุฏุณุชูโุง ุจุฑุง ุฑุณู ูููุฏุงุฑ
    target_col (str): ูุงู ุณุชูู ูุฏู ุจุฑุง ููุงุด hue (ูพุดโูุฑุถ 'Churn Label')
    """
    n_rows, n_cols = 2, 2
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, 10))
    axes = axes.flatten()

    for i, col in enumerate(categorical_cols):
        ax = axes[i]
        sns.countplot(data=df, x=col, hue=target_col, ax=ax, palette={'Yes': 'red', 'No': 'blue'})
        ax.set_title(f'{col} vs {target_col}\nMissing: {df[col].isna().sum()}')
        ax.set_xlabel(col)
        ax.set_ylabel('Count')
        ax.tick_params(axis='x', rotation=45)
        # ููุงุด ุชุนุฏุงุฏ ู ุฏุฑุตุฏ ุฑู ุณุชููโูุง
        total = len(df[col].dropna())
        for container in ax.containers:
            labels = [f'{int(v.get_height())} ({v.get_height()/total*100:.1f}%)' if v.get_height() > 0 else '' for v in container]
            ax.bar_label(container, labels=labels, fontsize=10, label_type='edge')

    # ุญุฐู ุณุงุจโูพูุงุชโูุง ุงุถุงู (ุฏุฑ ุตูุฑุช ูุงุฒ)
    for j in range(len(categorical_cols), len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()
````

````python
# TODO: ุขูุงุฏู ุณุงุฒ ุฏุงุฏู ูุง ุจุฑุง ุงุณุชูุงุฏู ุงุฒ ColumnTransformer

categorical_cols = X.select_dtypes(include=["object"]).columns.tolist()
numerical_cols = X.select_dtypes(include=["int64", "float64"]).columns.tolist()

# ุงุฌุงุฏ ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ("num", Pipeline([
            ('imputer', SimpleImputer(strategy='mean')),  # ูพุฑ ฺฉุฑุฏู ููุงุฏุฑ ฺฏูุดุฏู ุจุง ูุงูฺฏู
            ('scaler', StandardScaler())  # ูุฑูุงูโุณุงุฒ ุณุชููโูุง ุนุฏุฏ
        ]), numerical_cols),  # ุณุชููโูุง ุนุฏุฏ
        ("cat", Pipeline([
            ('imputer', SimpleImputer(strategy='most_frequent')),  # ูพุฑ ฺฉุฑุฏู ููุงุฏุฑ ฺฏูุดุฏู ุจุง ูุฏ
            ('encoder', OneHotEncoder(handle_unknown="ignore"))  # OneHotEncoding ุจุฑุง ุณุชููโูุง ุฏุณุชูโุง
        ]), categorical_cols)  # ุณุชููโูุง ุฏุณุชูโุง
    ]
)
````
