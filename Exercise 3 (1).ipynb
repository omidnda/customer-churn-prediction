````python
# ویژگی های دسته ایی
categorical_cols = ['Country', 'State', 'Gender', 'Senior Citizen', 'Partner', 'Dependents', 'Phone Service',
 'Multiple Lines', 'Internet Service', 'Online Security', 'Online Backup', 'Device Protection', 'Tech Support',
 'Streaming TV', 'Streaming Movies', 'Contract', 'Paperless Billing', 'Payment Method', 'Churn Label']
````

````python
# TODO: متد زیر را برای نشان دادن بار پلات تکمیل و متغیر های ذکر شده را بررسی کنید.
import math
def plot_categorical_bars(df,
                        categorical_columns,
                        rot=None,  # پارامتر چرخش برای برچسب‌های محور X
                        bar_width=0.3,  # عرض میله‌ها
                        palette='dark',
                        figsize=(16, 6)):
    n_cols = 3  # تعداد ستون‌های subplot
    n_rows = math.ceil(len(categorical_columns) / n_cols)  # تعداد ردیف‌ها بر اساس تعداد ویژگی‌ها
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(figsize[0], figsize[1] * n_rows))
    axes = axes.flatten()  

    for i, col in enumerate(categorical_columns):
        ax = axes[i]
        sns.countplot(data=df, x=col, hue='Churn Label', palette=palette, ax=ax)
        ax.set_title(f'Distribution of {col}')
        ax.set_xlabel(col)
        ax.set_ylabel('Count')
        if col == 'Payment Method':
            ax.tick_params(axis='x', labelrotation=45)
       

    # اگر تعداد ساب‌پلات‌ها بیشتر از تعداد ویژگی‌ها بود
    for j in range(i+1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()
````

````python
# TODO: بررسی بارپلات
plot_categorical_bars(df, categorical_cols)
````

````python
# TODO: متد زیر را برای نشان دادن هیستوگرام تکمیل و متغیر های ذکر شده را بررسی کنید.
def hist_plot(df, numeric_columns,
              bins=30, rot=None,
              figsize=(14, 8)):
    """
    این تابع چندین متغیر عددی را به صورت هستوگرام نمایش می‌دهد.
    پارامترها:
    df : DataFrame -> دیتافریم ورودی
    numeric_columns : list -> لیست متغیرهای عددی
    bins : int -> تعداد سطل‌های (bins) نمودار
    rot : int -> مقدار چرخش برچسب‌های محور X
    figsize : tuple -> اندازه نمودار
    """
    n_cols = 2  # تعداد ستون‌های subplot
    n_rows = math.ceil(len(numeric_cols) / n_cols) 
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(figsize[0], figsize[1] * 2))
    axes = axes.flatten()  
    for i,col in enumerate(numeric_cols):
        df[col] = pd.to_numeric(df[col], errors='coerce')
        ax=axes[i]
        sns.histplot(df[col], bins=bins,kde=True,ax=ax, legend=True, color="skyblue"), 
        # sns.histplot(df[col].dropna(), bins=bins,kde=True,ax=ax, legend=True, color="skyblue"), 
        ax.set_title(f'Histogram of {col}\nMean: {df[col].mean():.2f}, Std: {df[col].std():.2f}')
        ax.set_xlabel(col)
        ax.set_ylabel('Frequency')
        ax.tick_params(axis='x', labelrotation=rot)
    for j in range(i+1, len(axes)):
        fig.delaxes(axes[j])
    
    plt.tight_layout()
    plt.show()
````

````python
# TODO: از تابع نوشته استفاده کنید و هیستوگرام ویژگی های عددی را رسم کنید.
hist_plot(df, numeric_cols)
````

````python
# متغیر های مشخص شده برای Kde Plot
numeric_cols = ['Monthly Charges', 'Total Charges', 'Tenure Months', 'CLTV',]
````

````python
# TODO: متد زیر را برای نشان دادن KDE Plot تکمیل و متغیر های ذکر شده را بررسی کنید.
def plot_numeric_distributions(df, numeric_columns, target_col='Churn Label'):
    """
    رسم نمودارهای KDE برای متغیرهای عددی با تفکیک بر اساس متغیر هدف.
    این تابع توزیع متغیرهای عددی را برای هر مقدار از متغیر هدف (مانند 'Churn Label')
    در یک شبکه ۲×۲ نمایش می‌دهد.
    پارامترها:
    df (DataFrame): دیتافریم شامل داده‌ها
    numeric_cols (list): لیست ستون‌های عددی برای رسم نمودار
    target_col (str): نام ستون هدف برای دسته‌بندی (پیش‌فرض 'Churn Label')
    """
    n_cols = 2  # تعداد ستون‌های subplot
    n_rows = math.ceil(len(numeric_cols) / n_cols) 
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(10,8))
    axes = axes.flatten() 
    palette = {'No': 'blue', 'Yes': 'red'}

    # رسم هر کدام از ستون‌های عددی
    for i, col in enumerate(numeric_cols):
        df[col] = pd.to_numeric(df[col], errors='coerce')
        ax = axes[i]
        sns.kdeplot(data=df, x=df[col], hue=target_col, ax=ax, fill=True, common_norm=False,warn_singular=False, palette=palette)
        ax.set_title(f'Distribution of {col} by {target_col}\nMissing: {df[col].isna().sum()}')
        ax.set_xlabel(col)
        ax.set_ylabel('Density')
        # افزودن میانگین برای هر گروه
        for hue_val in df[target_col].unique():
            mean_val = df[df[target_col] == hue_val][col].mean()
            ax.axvline(mean_val, linestyle='--', label=f'Mean {hue_val}: {mean_val:.2f}', alpha=0.5)
        ax.legend()


    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()
````

````python
# TODO: از تابع نوشته استفاده کنید و KDE Plot ویژگی های عددی را رسم کنید.
plot_numeric_distributions(df, numeric_cols)
````

````python
# TODO: متد زیر را تکمیل کنید برای بررسی ارتباط متغیرهای دسته‌ای با  با هدف.
def plot_categorical_subplots(df,
                              categorical_cols,
                              target_col='Churn Label'):
    """
    رسم نمودارهای Countplot برای متغیرهای دسته‌ای همراه با متغیر هدف.
    این تابع نمودارهای شمارشی را برای هر متغیر دسته‌ای مشخص‌شده در کنار متغیر هدف
    (مانند 'Churn Label') در یک شبکه ۲×۲ نمایش می‌دهد.
    پارامترها:
    df (DataFrame): دیتافریم شامل داده‌ها
    categorical_cols (list): لیست ستون‌های دسته‌ای برای رسم نمودار
    target_col (str): نام ستون هدف برای نمایش hue (پیش‌فرض 'Churn Label')
    """
    n_rows, n_cols = 2, 2
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, 10))
    axes = axes.flatten()

    for i, col in enumerate(categorical_cols):
        ax = axes[i]
        sns.countplot(data=df, x=col, hue=target_col, ax=ax, palette={'Yes': 'red', 'No': 'blue'})
        ax.set_title(f'{col} vs {target_col}\nMissing: {df[col].isna().sum()}')
        ax.set_xlabel(col)
        ax.set_ylabel('Count')
        ax.tick_params(axis='x', rotation=45)
        # نمایش تعداد و درصد روی ستون‌ها
        total = len(df[col].dropna())
        for container in ax.containers:
            labels = [f'{int(v.get_height())} ({v.get_height()/total*100:.1f}%)' if v.get_height() > 0 else '' for v in container]
            ax.bar_label(container, labels=labels, fontsize=10, label_type='edge')

    # حذف ساب‌پلات‌های اضافی (در صورت نیاز)
    for j in range(len(categorical_cols), len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()
````

````python
# لیست کل خدماتی که توسط شرکت ارائه می شود.
service_columns = ['Phone Service', 'Multiple Lines', 'Internet Service',
                   'Online Security', 'Online Backup', 'Device Protection',
                   'Tech Support', 'Streaming TV', 'Streaming Movies', 'Paperless Billing']

````

````python
# TODO: تعداد خدماتی که هر مشتری مشترک شده را محاسبه کنید.
def calculate_total_services(df):
    def count_services(row):
        count = 0
        for col in service_columns:
            val = row[col]
            if col == 'Internet Service':
                if val != 'No':
                    count += 1
            else:
                if val == 'Yes':
                    count += 1
        return count

    df['Total Service Subscribed'] = df.apply(count_services, axis=1)
    return df
calculate_total_services(df_copy)
````

        تشخیص مشتریانی که پتانسیل فروش افزایشی دارند بر اساس Churn Score و CLTV.تشخیص مشتریانی با پتانسیل فروش افزایشی (Upsell Potential). این ویژگی برای شناسایی مشتریانی است که هم احتمال ریزش بالایی دارند و هم ارزش طول عمر بالایی برای شرکت دارند. برای این کار، از دو ویژگی Churn Score و CLTV استفاده می‌کنیم. اگر Churn Score مشتری بیشتر از 50 باشد و CLTV آن از میانگین بیشتر باشد، برای این مشتری ویژگی Upsell Potential را 1 و در غیر این صورت 0 قرار می‌دهیم. مقدار 1 یا 0 باید در متغیر جدید به نام <code>Upsell Potential</code> ذخیره شود. این ویژگی می‌تواند به مدل شما کمک کند تا مشتریانی که پتانسیل خرید بیشتری دارند را شناسایی کند.

````python
# تشخیص مشتریانی که پتانسیل فروش افزایشی دارند
def add_upsell_flag(df):
    cltv_mean = df['CLTV'].mean()
    df['Upsell Potential'] = df.apply(
        lambda row: 1 if row['Churn Score'] > 50 and row['CLTV'] > cltv_mean else 0,
        axis=1
    )
    return df
add_upsell_flag(df_copy)
````

            <ul>
                <li>0 تا 33: گروه <strong>Low</strong> (احتمال ترک پایین)</li>
                <li>33 تا 66: گروه <strong>Medium</strong> (احتمال ترک متوسط)</li>
                <li>66 تا 100: گروه <strong>High</strong> (احتمال ترک بالا)</li>
            </ul>
        </li>
        <li>
            این دسته‌بندی به مدل کمک می‌کند تا الگوهای رفتاری مشتریان را بهتر درک کرده و احتمال ترک سرویس را به‌طور دقیق‌تری پیش‌بینی کند. ویژگی جدید به نام <strong>Churn Score Group</strong> در دیتافریم ذخیره می‌شود. (تکنیک های Discretization )
        </li>

<br>
<br>

````python
#TODO: دسته‌بندی مشتریان بر اساس Churn Score
def add_churn_score_group(df):
    def score_group(row):
        if row["Churn Score"] <= 33:
            return "Low"
        elif row["Churn Score"] > 33 and row["Churn Score"] <= 66:
            return "Medium"
        elif row["Churn Score"] > 66 and row["Churn Score"] <= 100:
            return "High"
    df["Churn Score Group"] = df.apply(score_group, axis=1)
    return df

add_churn_score_group(df_copy)
````

````python
# TODO: آماده سازی داده های برای استفاده از ColumnTransformer

categorical_cols = X.select_dtypes(include=["object"]).columns.tolist()
numerical_cols = X.select_dtypes(include=["int64", "float64"]).columns.tolist()

# ایجاد ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ("num", Pipeline([
            ('imputer', SimpleImputer(strategy='mean')),  # پر کردن مقادیر گمشده با میانگین
            ('scaler', StandardScaler())  # نرمال‌سازی ستون‌های عددی
        ]), numerical_cols),  # ستون‌های عددی
        ("cat", Pipeline([
            ('imputer', SimpleImputer(strategy='most_frequent')),  # پر کردن مقادیر گمشده با مد
            ('encoder', OneHotEncoder(handle_unknown="ignore"))  # OneHotEncoding برای ستون‌های دسته‌ای
        ]), categorical_cols)  # ستون‌های دسته‌ای
    ]
)
````

````python
# TODO: ایجاد دیکشنری از مدل های مورد نظر برای آموزش و ارزیابی اولیه
# TODO: آموزش داده با استفاده از پایپلاین
# TODO: ارزیابی داده ها روی مجموعه آموزش و تست

models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(n_estimators=100),
    'Gradient Boosting': GradientBoostingClassifier(),
    'XGBoost': XGBClassifier(),
    'AdaBoost': AdaBoostClassifier()
}


for model_name, model in models.items():
    # ساخت Pipeline برای هر مدل
    pipe = Pipeline(steps=[
        ("preprocess", preprocessor),  # انجام preprocessing روی داده‌ها
        ("classifier", model)  # مدل
    ])
    
    # آموزش مدل
    pipe.fit(X_train, y_train)
    
    # پیش‌بینی
    y_pred = pipe.predict(X_test)
    
    # ارزیابی مدل
    print(f"🔷 {model_name} Report:\n", classification_report(y_test, y_pred))
    print("="*50)
````
